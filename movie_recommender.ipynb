```python
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    movie_data[feature] = movie_data[feature].apply(literal_eval)
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than 3 elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []
```

```python
movie_data['director'] = movie_data['crew'].apply(get_director)
```

```python
movie_data['genres']
```

```python
movie_data['director']
```

```python
fin_features=['cast', 'keywords','genres']
for feature in fin_features:
  movie_data[feature]=movie_data[feature].apply(get_list)
```

```python
movie_data.head()
```

```python
def clean_data(x):
    if isinstance(x,list):
        return[str.lower(i.replace(" ","")) for i in x]
    elif isinstance(x,str):
        return str.lower(x.replace(" ",""))
    else:
        return ""
```

```python
features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    movie_data[feature] = movie_data[feature].apply(clean_data)
```

```python
movie_data['director']
```

```python
def create_soup(x):
    return " ".join(x['cast']) + " " + " ".join(x['keywords']) + " " + " ".join(x['director']) + " " + " ".join(x['genres'])
```

```python
movie_data['soup'] = movie_data.apply(create_soup,axis = 1)
```

```python
movie_data['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words = 'english')
count_matrix = count.fit_transform(movie_data['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity = cosine_similarity(count_matrix,count_matrix)
```

```python
cosine_similarity.shape
```

```python
movie_data = movie_data.reset_index()
```

```python
movie_titles = pd.Series(movie_data.index, index=movie_data['title'])
```

```python
movie_titles
```

```python
def get_recs(title,cosine_similarity = cosine_similarity):
    title = movie_titles[title]
    sim_scores = list(enumerate(cosine_similarity[title]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:16]
    movie_indices = [i[0] for i in sim_scores]
    sim_scores
    return movie_data['title'].iloc[movie_indices]
```
